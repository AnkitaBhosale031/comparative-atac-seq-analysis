# Step 3: Import libraries

```python
import gseapy as gp
import pandas as pd
import matplotlib.pyplot as plt
```

# tep 4: Run GO Biological Process enrichment

```python
go_results = gp.enrichr(
    gene_list=genes,
    gene_sets="GO_Biological_Process_2023",
    organism="human",
    outdir=None
)

go_table = go_results.results

go_table.head(10)
```

# Step 5: Save results

```python
go_table.to_csv(
    "/home/ankita/chromatin_results/GO_BP_results.csv",
    index=False
)
```

# GO Dot Plot GO BP results

```python
import matplotlib.pyplot as plt

top_go = go_table.head(10)

plt.figure(figsize=(8,6))

plt.scatter(
    top_go["Combined Score"],
    top_go["Term"]
)

plt.xlabel("Combined Score")
plt.ylabel("GO Biological Process")
plt.title("Top Enriched GO Biological Processes")

plt.tight_layout()
plt.show()
```

# Bar plot

```python
plt.figure(figsize=(8,6))

plt.barh(
    top_go["Term"],
    top_go["Overlap"].str.split("/").str[0].astype(int)
)

plt.xlabel("Number of Genes")
plt.ylabel("GO Biological Process")
plt.title("Top GO Biological Processes")

plt.gca().invert_yaxis()

plt.tight_layout()

plt.savefig(
    "/home/ankita/chromatin_results/GO_BP_BarPlot.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

# kegg

```python
top_kegg = kegg_table.sort_values("Adjusted P-value").head(10).copy()

top_kegg["Gene Count"] = top_kegg["Overlap"].apply(lambda x: int(x.split("/")[0]))

plt.figure(figsize=(10,6))

plt.scatter(
    top_kegg["Combined Score"],
    top_kegg["Term"],
    s=top_kegg["Gene Count"]*20
)

plt.xlabel("Combined Score")
plt.ylabel("KEGG Pathway")
plt.title("Top 10 KEGG Pathways")

plt.tight_layout()

plt.savefig(
    "/home/ankita/chromatin_results/KEGG_DotPlot.png",
    dpi=300
)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.barh(
    top_kegg["Term"],
    top_kegg["Gene Count"]
)

plt.xlabel("Gene Count")
plt.ylabel("KEGG Pathway")
plt.title("Top 10 KEGG Pathways")

plt.gca().invert_yaxis()

plt.tight_layout()

plt.savefig(
    "/home/ankita/chromatin_results/KEGG_BarPlot.png",
    dpi=300
)

plt.show()
```
